# **Inheritance**

Inheritance is a fundamental concept in object oriented programming that allow a class to inherit attributes and methods froom another class.

**Inheritance** in Python allows one class (child/derived class) to **reuse the properties and methods** of another class (parent/base class).

Example:

Vehicle (parent class) has common features like start, stop.

Car and Bike (child classes) inherit from Vehicle and also have their own features.

```python
# inheritance
class car:
  def __init__(self,brand,model):
    self.brand=brand
    self.model=model
  def drive(self):
    print(f"the {self.brand} {self.model} is driving")

car1=car("Toyota","Camry")
car1.drive()

class Tesla(car):
  def __init__(self,model,is_selfdrive):
    super().__init__("Tesla",model)
    self.is_selfdrive=is_selfdrive
  def self_drive(self):
    if self.is_selfdrive:
      print(f"the {self.brand} {self.model} is in self drive mode")
    else:
      print("the selfdriving is not allowed is tesla")
  def drive(self):
    print(f"the {self.brand} {self.model} {self.is_selfdrive} is driving")
Tesla1=Tesla("model 3",True)
Tesla1.drive()
```

# **Method Overriding**

Method overriding allows a child class to provide a specific implementaion of a method that is already define in its parent class.

```python
## base class
class animal:
  def speak(self):
    return "Sound of the animal"

class Dog(animal):
  def speak(self):
    return "Woof"
class Cat(animal):
  def speak(self):
    return "meow"

def animal_speak(animal):
  print(animal.speak())

dog=Dog()
cat=Cat()
animal_speak(dog)
animal_speak(cat )
animal_speak(dog)
```

**Polymorphism with Function and methods**

```python
## Base class
class Shape:
  def area(self):
    return "the area of the figure"
##Derived class 1
class Rectangle(Shape):
    def __init__(self,width,height):
      self.width=width
      self.height=height
    def area(self):
        return self.width * self.height
## Derived class 2
class Circle(Shape):
  def __init__(self,radius):
    self.radius=radius
  def area(self):
    return 3.14*self.radius**2

def shape_area(shape):
  print(f" the are is {shape.area()}")
rectangle=Rectangle(5,10)
circle=Circle(7)
shape_area(rectangle)
shape_area(circle)
```

**Polymorphism with Abstract Base Classes**

Abstract Base classes are used to define common methods for a  group of related objects.They can enforce that derived classes implement particular methods,propoting consistency across defferent inmplementation.

```python
from abc import ABC,abstractmethod

## Define an abstruct class
class Vehicle(ABC):
  @abstractmethod
  def start_engine(self):
    pass
##Derived class 1
class Car(Vehicle):
  def start_engine(self):
    return "car engine started"

## Derived class 2
class Motorcycle(Vehicle):
  def start_engine(self):
    return "motorcycle engine started"

def vehicle_engine_start(vehicle):
  print(vehicle.start_engine())

car=Car()
motorcycle=Motorcycle()
vehicle_engine_start(car)
```

# **Encapsulation**

**Encapsulation** means ***binding*** or **wrapping** data (variables) and **methods** (functions) into a **single unit** (class).

It also provides a way to **restrict direct access** to data by controlling visibility (public, protected, private).

The main goal is data ***security*** and **controlled access.**

Think of it like a ***capsule*** that keeps **medicine** safe inside.

# **Access Modifiers in Python (by naming convention)**

1.**Public** → Accessible everywhere. (Default)

    self.name


```python
## public variable or access modifiers

class Person:
  def __init__(self,name,age):
    self.name=name             ## public variable
    self.age=age               ## public variable
def get_name(Person):
  return Person.name

person=Person("mahesh" ,32)
print(get_name(person))
```

```python
dir(Person)
```

2. **Private** → Cannot be accessed directly outside the class (name mangling is used).

    self.__salary


```python
from collections.abc import AsyncGenerator
class Person:
  def __init__(self,name,age):
    self.__name=name             ## private variable
    self.__age=age
def get_name(Person):
  return Person.__name

person=Person("mahesh" ,32)
print(get_name(person))
```

```python
dir(person)
```

3. **protected** → Meant to be used within the class and subclasses (not strictly enforced)

    self._age


```python
class Person:
  def __init__(self,name,age):
    self._name=name             ## protected vaeriable
    self._age=age               ## protected variable
class Employee(Person):
  def __init__(self,name,age,salary):
    super().__init__(name,age)
    self.salary=salary

employee=Employee("mahesh",32,50000)
print(employee._name)
```

**Encapsulation with getter and setter**

```python
class Person:
  def __init__(self,name,age):
    self.__name=name
    self.__age=age

## getter method for name
  def get_name(self):
    return self.__name

## setter method for name
  def set_name(self,name):
    self.__name=name

  def get_age(self):
    return self.__age

  def set_age(self,age):
    if age<0:
      print("age cannot be negative")
    else:
      self.__age=age

person=Person("mahesh",32)

print(person.get_name())
print(Person.get_age())

person.set_age(32)
print(person.get_name())
print(person.get_age())
```

# **Abstraction**

Abstraction is the concept of hiding the internal implementation details and showing only the necessary features of an object.this helps in reducing programming complexity and effort

***Abstraction*** means **hiding the internal implementation details** of a program and **showing only the necessary features to the user.**
It helps in reducing complexity and increases code reusability.

**Real-life Example**

When you drive a car —
you only use the **steering wheel, brakes, and accelerator,**
but you don’t know (or need to know) how the engine works internally.
That’s abstraction!

Abstraction is mainly achieved using **Abstract Classes** and **Abstract Methods.**

```python
from abc import ABC, abstractmethod

# Abstract base class
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")

    @abstractmethod
    def start_engine(self):
        pass

# Subclass implementing abstract method
class Car(Vehicle):
    def start_engine(self):
        print("The car engine is started")

# Function to operate any vehicle object
def operate_vehicle(v):
    v.start_engine()
    v.drive()

# Create Car object
car = Car()

# Operate the car
operate_vehicle(car)
```

# **OperatorOverloding**

Operator overloading means giving special meaning to operators (like +, -, *, /, >, ==, etc.) when they are used with user-defined objects (classes).

```python
class Vector:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def __add__(self,other):
    return Vector(self.x+other.x,self.y+other.y)

  def __sub__(self,other):
    return Vector(self.x-other.x,self.y-other.y)

  def __str__(self):
    return f"({self.x},{self.y})"

  def __mul__(self,other):
    return Vector(self.x*other.x,self.y*other.y)

  def __eq__(self,other):
    return self.x==other.x and self.y==other.y

v1=Vector(1,2)
v2=Vector(3,4)
print(v1+v2)
print(v1-v2)
print(v1*v2)
print(v1==v2)

```

```python
 def my_decorator(func):
    def wrapper():
        print("mahesh and sejal are the best couple in the their village")
        func()
        print("They both are married on 2028 in village")
    return wrapper

@my_decorator
def my_function():
    print("This is my function")

my_function()
```

```python
 # Decoratoe with arguments
 def repeat(n):
     def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
     return decorator

@repeat(3)
def greet(name):
    print(f"Hello, {name}!")

greet("mahesh")
```

```python
import threading
import time

def print_numbers():
  for i in range(5):
    time.sleep(2)
    print(f"Number: {i}")

def print_letter():
  for letter in "abcde":
    time.sleep(2)
    print(f"Letter: {letter}")

## creating a thread
t1=threading.Thread(target=print_numbers)
t2=threading.Thread(target=print_letter)

t1.start()
t2.start()

t1.join()
t2.join()

finished_time=time.time()
print(f"finished in {finished_time}")
```

# **MultiProcessing**

 A process is a **separate instance** of Python — it runs independently,

process that run in parallel

1 Cpu-Bound-Tasks that are heavy on cpu usage(e.g , mathamatical computatuions)

2 parallel executiojn-multiple cores of the cpu

```python
import multiprocessing
import time

def square_numbers():
    for i in range(5):
        time.sleep(2)
        print(f"square: {i * i}")

def cube_numbers():
    for i in range(5):
        time.sleep(2)
        print(f"cube: {i * i * i}")

if __name__ == "__main__":  # ✅ Corrected
    # Create two processes
    p1 = multiprocessing.Process(target=square_numbers)
    p2 = multiprocessing.Process(target=cube_numbers)

    t = time.time()

    p1.start()
    p2.start()

    p1.join()
    p2.join()

    finished_time = time.time() - t
    print(f"Finished in {finished_time:.2f} seconds")
```

# ***Memory management in python***

**Memeory management in python** involves a combination of automatic g**arbage collection,reference counting,and various internal** optimazation to effeciently manage memory allocation and deallocation

**Memory management** is how Python **allocates, tracks, and frees up memory (RAM)** used by variables, objects, and data structures while a program runs.

**In simple terms:**

When you create a variable or object in Python, it gets space in memory.
When it’s no longer needed, Python automatically removes it to free memory.

This automatic cleanup is done by **Python’s memory manager and garbage collector.**

***Reference Counting**

Reference counting is the primary method pyhton uses to manage memory.Each object in python maintains a count of references pointing to it. When a reference count drops to zero,ther memory occupied by the object is ceallocation

Python uses a system called **Reference** Counting to track how many variables refer to the same object.

```python
import sys

a=[]
print(sys.getrefcount(a))
```

```python
b=a
print(sys.getrefcount(b))
```

```python
del b
print(sys.getrefcount(a))
```

🧠 **Python keeps a reference count:**

When a new variable points to an object → count increases.

When a variable is deleted or reassigned → count decreases.

When count = 0 → object is automatically deleted from memory.
